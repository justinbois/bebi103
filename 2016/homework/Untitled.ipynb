```python
import numpy as np
```

```python
def log_prior(p):
    mu1, sigma1, mu2, sigma2, mu, sigma = p
    
    return -np.log(sigma) - np.log(sigma1) - np.log(sigma2) \
            - np.log(sigma) - (mu1 - mu)**2 / 2 / sigma**2 \
            - np.log(sigma) - (mu2 - mu)**2 / 2 / sigma**2    
```
